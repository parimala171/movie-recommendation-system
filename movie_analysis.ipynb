```python
#1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

plt.rcParams['figure.figsize'] = (8,5)
sns.set_style("whitegrid")
```

```python
#2. Load Dataset
df = pd.read_csv("movies.csv", low_memory=False)

df.head()
```

```python
#3. Basic Info
df.shape
df.info()
df.describe()
```

```python
#4. Select Required Columns
df = df[['title', 'overview', 'popularity', 'vote_average']]
```

```python
#5. Duplicate Removal
df.duplicated().sum()

df = df.drop_duplicates()

df.duplicated().sum()
```

```python
#6. Missing Values
df.isnull().sum()

df = df.dropna()
```

```python
#7. EDA 
#7.1 Popularity Distribution
sns.histplot(df['popularity'], bins=30)
plt.title("Popularity Distribution")
plt.show()
```

```python
#7.2 Rating Distribution
sns.histplot(df['vote_average'], bins=20)
plt.title("Rating Distribution")
plt.show()
```

```python
#7.3 Top Rated Movies
top_movies = df.sort_values(by='vote_average', ascending=False).head(10)

top_movies[['title', 'vote_average']]
```

```python
#8. Text Processing 
df['overview'] = df['overview'].str.lower()
```

```python
#9. Vectorization (NLP) 
cv = CountVectorizer(max_features=5000, stop_words='english')

vectors = cv.fit_transform(df['overview']).toarray()
```

```python
#10. Similarity Matrix
similarity = cosine_similarity(vectors)
```

```python
#11. Recommendation Function 
def recommend(movie):
    if movie not in df['title'].values:
        return ["Movie not found"]
    
    index = df[df['title'] == movie].index[0]
    distances = similarity[index]
    
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    recommendations = []
    for i in movie_list:
        recommendations.append(df.iloc[i[0]].title)
    
    return recommendations
```

```python
#12. Test Recommendation
recommend("Avatar")
```

```python
#13. Save Cleaned Data
df.to_csv("cleaned_data.csv", index=False)
```
